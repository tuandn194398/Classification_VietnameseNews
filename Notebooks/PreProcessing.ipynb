```python
# Hm tch t
# Tham s: folder_path - ng dn n folder cha cc file cn tch t
def get_data(folder_path):
    X = []
    y = []
    dirs = os.listdir(folder_path)
    for path in tqdm(dirs):
        file_paths = os.listdir(os.path.join(folder_path, path))
        for file_path in tqdm(file_paths):
            with open(os.path.join(folder_path, path, file_path), 'r', encoding="utf-16") as f:
                lines = f.readlines()
                lines = ' '.join(lines)
                lines = gensim.utils.simple_preprocess(lines)
                lines = ' '.join(lines)
                lines = word_tokenize(lines, format="text")
                #lines = ViTokenizer.tokenize(lines)
                X.append(lines)
                y.append(path)

    return X, y
```

```python
# Thc hin tch t i vi tp train
X_data, y_data = get_data('/content/drive/MyDrive/Project 3/new train')
```

```python
# Lu tp train  tch t vo file .pkl
import pickle

pickle.dump(X_data, open('/content/drive/MyDrive/Project 3/X_data.pkl', 'wb'))
pickle.dump(y_data, open('/content/drive/MyDrive/Project 3/y_data.pkl', 'wb'))
```
